```python
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_formats = ['svg']
```

```python
rnn = np.load("rnn_train_loss.npy")
gru = np.load("gru_train_loss.npy")
lstm = np.load("lstm_train_loss.npy")
rnn_att = np.load("rnn_att_train_loss.npy")
gru_att = np.load("gru_att_train_loss.npy")
lstm_att = np.load("lstm_att_train_loss.npy")
fine_tune = np.load("fine_tune_train_loss.npy")
```

```python
plt.plot(rnn, label="Vanilla RNN")
plt.plot(rnn_att, label="RNN + Attention")
plt.plot(gru, label="Vanilla GRU")
plt.plot(gru_att, label="GRU + Attention")
plt.plot(lstm, label="Vanilla LSTM")
plt.plot(lstm_att, label="LSTM + Attention")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid()
plt.legend()
plt.savefig("./report/figures/vanilla.pdf", bbox_inches='tight')
plt.show()
```

```python
plt.plot(fine_tune, label="Opus-mt fine tunning")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid()
plt.legend()
plt.savefig("./report/figures/fine_tunning.pdf", bbox_inches='tight')
plt.show()
```
